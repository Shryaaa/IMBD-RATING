```python
 #importing the modules
from bokeh.plotting import figure, output_file, show
import pandas as pd
  
  
```

```python
graph = figure(title = "IPL")
  
# reading the database
df= pd.read_csv("IPL Auction 2022.csv")
  
# plotting the graph
graph.vbar(df['Player'], top=df['Sold Price'], 
           legend_label = "sold price vs Player", color='green')
  
graph.vbar(df['Player'], top=df['New Franchise'], 
           legend_label = "2021 Team VS Base Price", color='red')
  
graph.legend.click_policy = "hide"
  
# displaying the model
show(graph)
```
